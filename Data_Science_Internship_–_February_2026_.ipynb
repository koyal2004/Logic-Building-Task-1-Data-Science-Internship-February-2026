```python
# =========================================
# Logic Building Task â€“ 1
# =========================================


# -----------------------------------------
# 1. User Login Check
# -----------------------------------------

print("----- Task 1: User Login Check -----")

stored_username = "admin"
stored_password = "1234"

entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

if entered_username == stored_username and entered_password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")

print("\n")


# -----------------------------------------
# 2. Pass / Fail Analyzer
# -----------------------------------------

print("----- Task 2: Pass / Fail Analyzer -----")

marks = [45, 78, 90, 33, 60]

passed = 0
failed = 0

for score in marks:
    if score >= 50:
        passed += 1
    else:
        failed += 1

print("Total Students:", len(marks))
print("Number of Students Passed:", passed)
print("Number of Students Failed:", failed)

print("\n")


# -----------------------------------------
# 3. Simple Data Cleaner
# -----------------------------------------

print("----- Task 3: Simple Data Cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for n in names:
    cleaned_value = n.strip().lower()
    cleaned_names.append(cleaned_value)

print("Cleaned Names List:", cleaned_names)

print("\n")


# -----------------------------------------
# 4. Message Length Analyzer
# -----------------------------------------

print("----- Task 4: Message Length Analyzer -----")

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    message_length = len(msg)
    print(f"Message: '{msg}' | Length: {message_length}")

    if message_length > 10:
        print("Status: Long message detected")

print("\n")


# -----------------------------------------
# 5. Error Message Detector
# -----------------------------------------

print("----- Task 5: Error Message Detector -----")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_total = 0

for entry in logs:
    if entry == "ERROR":
        error_total += 1

print("Total ERROR entries found:", error_total)

print("\n")
print("All tasks executed successfully.")
```
